# **🧠Models**

## **1) STT Model**

```python
processor_STT = AutoProcessor.from_pretrained("openai/whisper-medium")
model_STT = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-medium")
```

```python
def transcribe_audio(audio_input):
    inputs = processor_STT(audio_input, return_tensors="pt", sampling_rate=16000)

    forced_decoder_ids = processor_STT.get_decoder_prompt_ids(
        language="en", task="transcribe"
    )

    with torch.no_grad():
        predicted_ids = model_STT.generate(
            inputs.input_features,
            forced_decoder_ids=forced_decoder_ids
        )

    transcription = processor_STT.batch_decode(predicted_ids, skip_special_tokens=True)[0]
    return transcription
```

## **2) Language Model**

```python
tokenizer_LLM = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model_LLM = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
```

```python
chat_history_ids = None

def generate_reply(user_input):
    global chat_history_ids
    new_input_ids = tokenizer_LLM.encode(user_input + tokenizer_LLM.eos_token, return_tensors='pt')

    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        bot_input_ids = new_input_ids

    # Generate response
    chat_history_ids = model_LLM.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer_LLM.eos_token_id,
        do_sample=True,
        top_k=50,
        top_p=0.95
    )

    response = tokenizer_LLM.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response
```

## **3) TTS**

```python
def text_to_speech(text,lang_code):
  pipeline = KPipeline(lang_code=lang_code)                  #a,b for english and e for spanish and f for french
  generator = pipeline(text, voice='af_heart')               # af_ -> sarah , nicole , river , bella
  for i, (gs, ps, audio) in enumerate(generator):
      print(i, gs, ps)
      display(Audio(data=audio, rate=24000, autoplay=i==0))
      sf.write(f'{i}.wav', audio, 24000)
```

## **Fast chat Function**

```python
chat_history_ids = None  # Keep this globally at the top of your script

def fast_chat():
    global recorded_audio_bytes

    # Callback from JavaScript to Python
    def handle_audio_recording(base64_audio):
        global recorded_audio_bytes, chat_history_ids

        # Decode audio from base64
        recorded_audio_bytes = base64.b64decode(base64_audio)
        print("✅ Recording complete. Processing audio...")

        # Load and process audio
        audio = AudioSegment.from_file(io.BytesIO(recorded_audio_bytes))
        sample_rate = audio.frame_rate

        # Convert to numpy and normalize
        audio_np = np.array(audio.get_array_of_samples()).astype(np.float32) / (2**15)
        if audio.channels == 2:
            audio_np = audio_np.reshape((-1, 2)).mean(axis=1)  # Convert stereo to mono

        display(Audio(data=audio_np, rate=sample_rate))

        # Resample to 16k for Whisper
        audio_input = librosa.resample(audio_np, orig_sr=sample_rate, target_sr=16000)
        audio_input = audio_input * (10**-5)

        # Transcribe
        print("Working on the transcription part...")
        transcription = transcribe_audio(audio_input)
        print("👤 You:", transcription)

        # Exit if user says goodbye
        if transcription.strip().lower() in ["exit", "quit", "bye"]:
            print("🤖 Bot: 👋 See you soon, Goodbye!")
            text_to_speech("See you soon, Goodbye", "a")
            return  # ✅ Exit without recalling fast_chat

        # Generate reply using LLM
        reply = generate_reply(transcription)
        print("🤖 Bot:", reply)

        # Speak the response
        text_to_speech(reply, "a")

        # ✅ Wait for user to press "Record" again manually

    def notify_recording_started():
        print("🎙 Recording started...")

    # Register Python callbacks to be invoked from JavaScript
    output.register_callback('notebook.handle_audio_recording', handle_audio_recording)
    output.register_callback('notebook.notify_recording_started', notify_recording_started)

    # Display recording buttons and JS
    display(HTML('''
      <button onclick="startRecording()">🎙 Start Recording</button>
      <button onclick="stopRecording()">⏹ Stop Recording</button>
      <p><small>Allow microphone access when prompted.</small></p>
      <script>
        let recorder, stream;
        let audioChunks = [];

        async function startRecording() {
          stream = await navigator.mediaDevices.getUserMedia({ audio: true });
          recorder = new MediaRecorder(stream);
          audioChunks = [];

          recorder.ondataavailable = e => audioChunks.push(e.data);
          recorder.onstop = async () => {
            const blob = new Blob(audioChunks);
            const arrayBuffer = await blob.arrayBuffer();
            const base64Audio = btoa(String.fromCharCode(...new Uint8Array(arrayBuffer)));
            google.colab.kernel.invokeFunction('notebook.handle_audio_recording', [base64Audio], {});
          };

          recorder.start();
          google.colab.kernel.invokeFunction('notebook.notify_recording_started', [], {});
        }

        function stopRecording() {
          recorder.stop();
          stream.getTracks().forEach(track => track.stop());
        }
      </script>
    '''))

    print("🎤 Press 'Start Recording', then 'Stop Recording' when you're done...")
```

```python
%%writefile app.py
# ===== ABSOLUTELY FIRST COMMANDS =====
import streamlit as st

# THIS MUST BE THE VERY FIRST STREAMLIT COMMAND
st.set_page_config(
    page_title="Voice Chatbot",
    page_icon="🎙️",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Now we can do other imports and setup
import os
os.environ["STREAMLIT_SERVER_ENABLE_FILE_WATCHER"] = "false"
os.environ["TOKENIZERS_PARALLELISM"] = "false"

import warnings
warnings.filterwarnings("ignore", message=".*cuFFT.*")
warnings.filterwarnings("ignore", message=".*cuDNN.*")
warnings.filterwarnings("ignore", message=".*cuBLAS.*")
warnings.filterwarnings("ignore", message="Some weights of WhisperForConditionalGeneration were not initialized")

import streamlit.watcher
streamlit.watcher.local_sources_watcher.get_module_paths = lambda _: []

import soundfile as sf
import numpy as np
from PIL import Image
import time
from streamlit_webrtc import webrtc_streamer, WebRtcMode
import av
import queue
from datetime import datetime
import torch
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq, AutoModelForCausalLM, AutoTokenizer
from kokoro import KPipeline

@st.cache_resource(show_spinner="Loading speech-to-text model...")
def load_stt_model():
    processor = AutoProcessor.from_pretrained("openai/whisper-medium")
    model = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-medium")
    return processor, model

@st.cache_resource(show_spinner="Loading language model...")
def load_llm_model():
    tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
    model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
    return tokenizer, model

@st.cache_resource(show_spinner="Loading text-to-speech pipelines...")
def load_tts_pipelines():
    return {
        'english': KPipeline(lang_code='a'),
        'spanish': KPipeline(lang_code='e'),
        'french': KPipeline(lang_code='f')
    }

with st.spinner("Loading AI models... Please wait"):
    processor_STT, model_STT = load_stt_model()
    tokenizer_LLM, model_LLM = load_llm_model()
    tts_pipelines = load_tts_pipelines()
    chat_history_ids = None

def transcribe_audio(audio_input):
    inputs = processor_STT(audio_input, return_tensors="pt", sampling_rate=16000)
    forced_decoder_ids = processor_STT.get_decoder_prompt_ids(language="en", task="transcribe")

    with torch.no_grad():
        predicted_ids = model_STT.generate(
            inputs.input_features,
            forced_decoder_ids=forced_decoder_ids
        )

    return processor_STT.batch_decode(predicted_ids, skip_special_tokens=True)[0]

def generate_reply(user_input):
    global chat_history_ids
    new_input_ids = tokenizer_LLM.encode(user_input + tokenizer_LLM.eos_token, return_tensors='pt')

    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        bot_input_ids = new_input_ids

    chat_history_ids = model_LLM.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer_LLM.eos_token_id,
        do_sample=True,
        top_k=50,
        top_p=0.95
    )

    return tokenizer_LLM.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

def text_to_speech(text, lang_code, voice):
    pipeline = tts_pipelines[lang_code.lower()]
    generator = pipeline(text, voice=f"af_{voice.lower()}")
    for i, (gs, ps, audio) in enumerate(generator):
        return audio

SAMPLE_RATE = 44100
CHANNELS = 1
AUDIO_FORMAT = 'WAV'
AUDIO_SUBTYPE = 'PCM_16'

col1, col2 = st.columns([1, 3])
with col1:
    st.image("https://cdn-icons-png.flaticon.com/512/4712/4712035.png", width=100)
with col2:
    st.title("🎙️ Voice Interactive Chatbot")
    st.caption("Speak, listen, and chat with our AI assistant")

if 'messages' not in st.session_state:
    st.session_state.messages = []
if 'audio_queue' not in st.session_state:
    st.session_state.audio_queue = queue.Queue()

with st.sidebar:
    st.header("Settings")
    language = st.selectbox("Select Language", ("English", "Spanish", "French"), index=0)
    voice = st.selectbox("Select Voice", ("Sarah", "Nicole", "River", "Bella"), index=0)
    st.markdown("---")
    st.markdown("### About")
    st.info("This is an interactive voice chatbot that can understand and respond to your voice commands.")

chat_container = st.container()

with chat_container:
    for message in st.session_state.messages:
        with st.chat_message(message["role"]):
            st.markdown(message["content"])
            if message.get("audio"):
                st.audio(message["audio"], format='audio/wav')

class AudioProcessor:
    def __init__(self):
        self.frames = []

    def recv(self, frame: av.AudioFrame) -> av.AudioFrame:
        audio = frame.to_ndarray()
        if audio.ndim == 2:
            audio = np.mean(audio, axis=1)
        audio = (audio * (2**15 - 1)).astype(np.int16)
        self.frames.append(audio)
        return frame

    def get_audio(self):
        if self.frames:
            return np.concatenate(self.frames)
        return None

    def reset(self):
        self.frames = []

RTC_CONFIGURATION = {
    "iceServers": [
        {"urls": ["stun:stun.l.google.com:19302"]},
        {"urls": ["stun:stun1.l.google.com:19302"]},
        {"urls": ["stun:stun2.l.google.com:19302"]}
    ]
}

st.markdown("### Speak to the chatbot")
webrtc_ctx = webrtc_streamer(
    key="voice-chatbot",
    mode=WebRtcMode.SENDONLY,
    audio_processor_factory=AudioProcessor,
    media_stream_constraints={"audio": True, "video": False},
    async_processing=True,
    rtc_configuration=RTC_CONFIGURATION
)

# Start Recording
if st.button("Start Recording"):
    if webrtc_ctx.audio_processor:
        webrtc_ctx.audio_processor.reset()
        st.session_state["recording_started"] = time.time()
        st.info("Recording... Speak now!")

# Process Recording
if st.button("Process Recording") and webrtc_ctx.audio_processor:
    elapsed = time.time() - st.session_state.get("recording_started", 0)
    if elapsed < 2:
        st.warning("Please speak for at least 2 seconds before processing.")
    else:
        audio_data = webrtc_ctx.audio_processor.get_audio()
        if audio_data is not None:
            try:
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                recording_file = f"user_recording_{timestamp}.wav"
                sf.write(recording_file, audio_data, SAMPLE_RATE, format=AUDIO_FORMAT, subtype=AUDIO_SUBTYPE)

                audio_input, sample_rate = sf.read(recording_file)
                audio_input = audio_input.astype(np.float32)

                with st.spinner("Transcribing..."):
                    user_input = transcribe_audio(audio_input)

                with chat_container:
                    with st.chat_message("user"):
                        st.markdown(user_input)
                        st.audio(recording_file)

                st.session_state.messages.append({"role": "user", "content": user_input, "audio": recording_file})

                with st.spinner("Generating response..."):
                    bot_response = generate_reply(user_input)
                    lang = language.lower()
                    audio = text_to_speech(bot_response, lang, voice)

                    response_file = f"bot_response_{timestamp}.wav"
                    sf.write(response_file, audio, 24000, format=AUDIO_FORMAT, subtype=AUDIO_SUBTYPE)

                    with chat_container:
                        with st.chat_message("assistant"):
                            st.markdown(bot_response)
                            st.audio(response_file)

                    st.session_state.messages.append({
                        "role": "assistant",
                        "content": bot_response,
                        "audio": response_file
                    })

            except Exception as e:
                st.error(f"Error processing audio: {e}")

# Text input fallback
user_text = st.chat_input("Or type your message here...")
if user_text:
    with chat_container:
        with st.chat_message("user"):
            st.markdown(user_text)

    st.session_state.messages.append({"role": "user", "content": user_text})

    with st.spinner("Generating response..."):
        bot_response = generate_reply(user_text)
        lang = language.lower()
        audio = text_to_speech(bot_response, lang, voice)

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        response_file = f"bot_response_{timestamp}.wav"
        sf.write(response_file, audio, 24000, format=AUDIO_FORMAT, subtype=AUDIO_SUBTYPE)

        with chat_container:
            with st.chat_message("assistant"):
                st.markdown(bot_response)
                st.audio(response_file)

        st.session_state.messages.append({
            "role": "assistant",
            "content": bot_response,
            "audio": response_file
        })
```
